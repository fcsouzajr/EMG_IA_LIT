```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import pandas as pd

# Função para aplicar janelas deslizantes em um sinal de EMG
def sliding_window_emg(signal, window_size):
    windows = []
    for i in range(len(signal) - window_size + 1):
        window = signal[i:i + window_size]
        windows.append(window)
    return np.array(windows)


# Coluna a ser utilizada (por exemplo, "amplitude")
emg_signal = dataset1["EMG1"].values

# Parâmetros da janela deslizante
window_size = 200  # Tamanho da janela

# Aplicar janelas deslizantes no sinal de EMG
emg_windows = sliding_window_emg(emg_signal, window_size)

# Normalizar os dados para a escala de cores
norm = mcolors.Normalize(vmin=emg_windows.min(), vmax=emg_windows.max())

# Visualizar o sinal original e as janelas deslizantes
plt.figure(figsize=(14, 6))

plt.subplot(2, 1, 1)
plt.plot(emg_signal, label='Sinal de EMG')
plt.title('Sinal de EMG a partir de Arquivo CSV')
plt.xlabel('Amostra')
plt.ylabel('Amplitude')
plt.legend()

plt.subplot(2, 1, 2)
plt.imshow(emg_windows, aspect='auto', cmap='viridis', extent=[0, len(emg_signal), 0, window_size], norm=norm)
plt.title('Janelas Deslizantes')
plt.xlabel('Amostra')
plt.ylabel('Índice da Janela')
plt.colorbar(label='Amplitude')
plt.tight_layout()

plt.show()
```
